Seminar session 1 and 2

```python
#User defining a function to find the outliers
def find_outliers(data_frame, column):
    q1 = data_frame[column].quantile(0.25)
    q3 = data_frame[column].quantile(0.75)
    IQR = q3 - q1
    lower_bound = q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR
    outliers = data_frame[(data_frame[column] < lower_bound) | (data_frame[column] > upper_bound)]
    return outliers
```

Seminar session 1 and 2

```python
#Finding outliers for 'Age' column
age_outliers = find_outliers(data_frame, 'Age')
print(f"Age outliers found: {len(age_outliers)}")
print(age_outliers['Age'])
```

Seminar session 1 and 2

```python
#Solution for 'Age' outliers
median_age = data_frame.loc[data_frame['Age'].between(30, 70), 'Age'].median()
data_frame['Age'] = data_frame['Age'].apply(lambda x: median_age if pd.isnull(x) or x > 70 or x < 30 else x)
```

Seminar session 1 and 2

```python
#Solution for 'Tumor_Size' outliers
median_tumor_size = data_frame.loc[data_frame['Tumor_Size'].between(1, 70), 'Tumor_Size'].median()
data_frame['Tumor_Size'] = data_frame['Tumor_Size'].apply(lambda x: median_tumor_size if pd.isnull(x) or x < 1 or x > 70 else x)
```

Seminar session 1 and 2

```python
#Solution for 'Regional_Node_Examined' outliers
median_RNE = data_frame.loc[data_frame['Regional_Node_Examined'].between(0, 34), 'Regional_Node_Examined'].median()
data_frame['Regional_Node_Examined'] = data_frame['Regional_Node_Examined'].apply(lambda x: median_RNE if pd.isnull(x) else min(x, 34))
data_frame['Regional_Node_Positive'] = data_frame['Regional_Node_Positive'].apply(lambda x: min(x, 34) if not pd.isnull(x) else x)
```

Seminar session 1 and 2

```python
#Solution for 'Survival_Months' outliers
median_survival_months = data_frame.loc[(data_frame['Survival_Months'] >= 5) & (data_frame['Survival_Months'] <= 107), 'Survival_Months'].median()
data_frame['Survival_Months'] = data_frame['Survival_Months'].apply(lambda x: median_survival_months if (x < 5) or (x > 107) else x)
```
